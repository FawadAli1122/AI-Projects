```python
train_db = concrete.sample(frac=0.7, random_state= 10)

test_db = concrete.drop(train_db.index)
```

```python
train_features = train_db.copy()
test_features = test_db.copy()

train_labels = train_features.pop('Concrete compressive strength(MPa, megapascals) ')
test_labels = test_features.pop('Concrete compressive strength(MPa, megapascals) ')

print(train_labels.head(), test_labels.head(), train_features.head(), test_features.head())
```

```python
import tensorflow as tf
from tensorflow.keras.layers.experimental import preprocessing
normalizer = preprocessing.Normalization()
normalizer.adapt(np.array(train_features))
print(normalizer.mean.numpy())
```

```python
train_features.describe().transpose()[['mean', 'std']]
```

```python
linear_model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error')
```

```python
history = linear_model.fit(
    train_features, train_labels, 
    # total Epochs equal to 100
    epochs=100,
    verbose=0,
    # Calculate validation results on 20% of the training data
    validation_split = 0.2)
history.history
```
