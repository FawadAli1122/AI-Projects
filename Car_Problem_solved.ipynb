```python
import numpy as np

num_train_data = train_data.drop(cars_categorical.columns, axis = 1)
num_test_data = test_data.drop(cars_categorical.columns, axis = 1)

mean = np.mean(num_train_data, axis = 0)
std = np.std(num_train_data, axis = 0)

num_train_data -= mean
num_train_data /= std

num_test_data -= mean
num_test_data /= std

train_data[num_train_data.columns] = num_train_data
test_data[num_train_data.columns] = num_test_data

normalized_data = pd.concat((train_data, test_data))
```

```python
normalized_data.head()
```

```python
data_fin = pd.get_dummies(normalized_data, columns = cars_categorical.columns)
```

```python
data_fin.head()
```

```python
train_data = data_fin.sample(frac=0.7, random_state=0)
test_data = data_fin.drop(train_data.index)
train_prices = target_prices.sample(frac=0.7, random_state=0)
test_prices = target_prices.drop(train_prices.index)
train_data.shape
```

```python
# Now making a model:

from keras import models, layers, regularizers

def build_model():
  model = models.Sequential()
  
  model.add(layers.Dense(10, activation = 'relu', kernel_regularizer = regularizers.l2(0.001), input_shape = (199,)))
  model.add(layers.Dropout(0.3))
  model.add(layers.Dense(8, activation = 'relu', kernel_regularizer = regularizers.l2(0.001)))
  model.add(layers.Dropout(0.3))
  model.add(layers.Dense(6, activation = 'relu', kernel_regularizer = regularizers.l2(0.001)))
  model.add(layers.Dropout(0.3))
  
  model.add(layers.Dense(1))


  model.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['mae'])
  return model
```

```python
model = build_model()
model.fit(train_data, train_prices,
          epochs = 100, batch_size = 8, verbose = 0)

test_mse , test_mae = model.evaluate(test_data, test_prices)

test_mae
```
